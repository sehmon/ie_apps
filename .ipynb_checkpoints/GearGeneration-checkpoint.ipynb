# Gear Generation

Process:
1. Run script to generate gear layout
2. Run save scripts to export individual gears, layout file, and master layout file (in IN)
3. Open master file in Illustrator, then export as svg (at 72 DPI)
4. Open SVG export in Deepnest (at 72 DPI)
5. Open Deepnest export in Fusion, scale by 1.333 (Fusion import is at 96 DPI)

---

## Imports and Constants

```python
import math
import random
import os
import json
import string
import svgwrite

from involute_gear import InvoluteGear
from mathutils import *

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from datetime import datetime

%matplotlib inline

# Define starting constants
width = 18
height = 24

axis_radius = 0.25
motor_shaft_radius = 0.19 / 2
target_n_gears = 9
```

## Class Definitions

```python
class Gear:
    generated_ids = set()
    
    def __init__(self, x, y, radius, label="New Gear", master=False):
        self.x = x
        self.y = y
        self.radius = radius
        self.label = label
        self.master=master
        self.id = self.generate_unique_id()

    @classmethod
    def generate_unique_id(cls):
        while True:
            new_id = ''.join(random.choices(string.ascii_uppercase + string.digits, k=4))
            if new_id not in cls.generated_ids:
                cls.generated_ids.add(new_id)
                return new_id
    
    def position(self):
        return (self.x, self.y)

    def to_dict(self):
        return {"x": self.x, "y": self.y, "radius": self.radius, "label": self.label, "id": self.id}

    def __repr__(self):
        return f"{"*" if self.master else ""}{self.id} ({self.label}) - pos:{self.x, self.y} radius: {self.radius}"

    def get_involute_gear(self):
        d = 2 * self.radius
        m = 0.2
        t = int(d / m)

        return InvoluteGear(teeth=t, module=m, max_steps=100)
```

## Helper Functions

```python
def get_distance(x1, y1, x2, y2):
    """
    """
    return math.sqrt((x2-x1)**2 + (y2-y1)**2)

def are_overlapping(x1, y1, r1, x2, y2, r2):
    """
    """
    return get_distance(x1, y1, x2, y2) < (r1+r2)

def out_of_bounds(x, y, r, w, h):
    """
    """
    return (x-r) < 0 or (y-r) < 0 or (x+r) > w or (y+r) > h

def parse_dimension(dim):
    """
    """
    return float(dim.split("in")[0].strip())

def initialize_plot(width, height):
    """
    """
    fig, ax = plt.subplots(figsize=(width, height), dpi=72)
    plt.axis('off')
    plt.margins(0, 0)
    plt.subplots_adjust(top=1, bottom=0, right=1, left=0, hspace=0, wspace=0)
    return fig, ax

def set_plot_parameters(ax, plt):
    """
    """
    ax.set_aspect('equal')
    ax.set_xlim(0, width)
    ax.set_ylim(0, height)
    
    plt.axis('off')

def save_gears_to_json(gears, output_folder_path, filename="gears.json"):
    """
    """
    with open(os.path.join(output_folder_path, filename), 'w') as json_file:
        json.dump([gear.to_dict() for gear in gears], json_file)


def prepare_output_folder():
    """
    Creates a timestamped output folder and returns its path.
    """
    output_parent_folder_path = "../gear_generations"
    todays_folder_name = datetime.now().strftime("%Y-%m%d")
    generation_folder_name = datetime.now().strftime("%H%M%S")

    full_path = os.path.join(
        output_parent_folder_path,
        todays_folder_name,
        generation_folder_name
    )
    
    os.makedirs(full_path, exist_ok=True)
    return full_path
```

## Main Logic

### Gear Generation

```python
# Generate Gears
master_gear = Gear(width/2, height/2, 2.0, "master", master=True)
gears = [master_gear]

while len(gears) < target_n_gears:
    new_r = round(random.uniform(1, 5), 1)
    angle = random.random() * 2 * math.pi
    connecting_gear = random.choice(gears)
    
    h = connecting_gear.radius + new_r
    
    x = round(h * math.cos(angle) + connecting_gear.x, 2)
    y = round(h * math.sin(angle) + connecting_gear.y, 2)

    valid_position = True
    for g in gears:
        if are_overlapping(x, y, new_r, g.x, g.y, g.radius) or out_of_bounds(x, y, new_r, width, height):
            valid_position = False
            break

    if valid_position:
        gears.append(Gear(x, y, new_r))
gears
```

### Plotting

```python
# Plot Gears
fix, ax = initialize_plot(width, height)

for gear in gears:
    g = gear.get_involute_gear()
    points_gear = np.dot(rotation_matrix(g.theta_tooth_and_gap / 2), g.generate_gear())
    plt.plot(gear.x + points_gear[0,:], gear.y + points_gear[1,:])
    
    circle_radius = motor_shaft_radius if gear.master else axis_radius
    ax.add_patch(patches.Circle((gear.x, gear.y), circle_radius, fill=False, color='black'))

set_plot_parameters(ax, plt)
plt.show()
```

### File Saving

```python
# Prep the output folder for generations
output_folder_path = prepare_output_folder()
save_gears_to_json(gears, output_folder_path, filename="gears.json")

# Save current batch of gears
fix, ax = initialize_plot(width, height)

# Generate hole layout
for gear in gears:
    circle_radius = motor_shaft_radius if gear.master else axis_radius
    ax.add_patch(patches.Circle((gear.x, gear.y), circle_radius, fill=False, color='black'))

set_plot_parameters(ax, plt)
plt.savefig(os.path.join(output_folder_path, "layout_holes.svg"), format="svg", dpi=72)

master_file = svgwrite.Drawing(size=(f"{len(gears) * 100}in", "800px"))  # Use pixels for better compatibility
x_offset = 0

# Place gears
for gear in gears:
    gear_group = master_file.g(transform=f"translate({x_offset}, 0)")
    
    # Generate Gear
    g = gear.get_involute_gear()
    
    points_gear = np.dot(rotation_matrix(g.theta_tooth_and_gap / 2), g.generate_gear())
    plt.plot(gear.x + points_gear[0,:], gear.y + points_gear[1,:])
    
    gear_svg = g.get_svg(svgwrite.inch)
    
    gear_group.add(gear_svg)

    gear_width = parse_dimension(gear_svg['width'])
    gear_height = parse_dimension(gear_svg['height'])
    
    # Generate axis
    circle_radius = motor_shaft_radius if gear.master else axis_radius
    gear_group.add(gear_svg.circle((gear_width / 2 * svgwrite.inch, gear_height / 2 * svgwrite.inch), r=circle_radius * svgwrite.inch, fill='red'))
    gear_svg.add(gear_svg.circle((0, 0), r=circle_radius, fill='red'))

    
    # Add label
    gear_group.add(gear_svg.text(
        gear.id, 
        insert=(gear_width / 2 * svgwrite.inch, gear_height / 2 * svgwrite.inch), 
        text_anchor="middle",  # Center the text horizontally
        alignment_baseline="middle",  # Center the text vertically
        font_size="0.25in",  # Adjust font size
        fill="blue"  # Text color
    ))

    
    gear_svg.saveas(os.path.join(output_folder_path, f'{gear.id}.svg'))
    master_file.add(gear_group)
    
    x_offset += 3 * 75 * gear.radius

master_file.saveas(os.path.join(output_folder_path, f'all_gears.svg'))

# Save final layout as svg
set_plot_parameters(ax, plt)
plt.savefig(os.path.join(output_folder_path, "layout_gears.svg"), format="svg", dpi=72)
```
